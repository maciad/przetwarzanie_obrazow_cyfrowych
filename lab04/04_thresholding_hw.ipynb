Pokazana w ramach podstawowej części ćwiczenia binaryzacja adaptacyjna działa dobrze, ale jest dość złożona obliczeniowo (choć oczywiście należy mieć świadomość, że implementację można zoptymalizować i wyeliminować pewne powtarzające się obliczenia).

Aby je wyeliminować należy zastosować interpolację, co zostanie szczegółowo opisane poniżej.
Warto zaznaczyć, że podobny mechanizm interpolacji stosowany jest w poznanym wcześniej algorytmie CLAHE.
Zainteresowane osoby odsyłam do artykułu na [Wikipedii](https://en.wikipedia.org/wiki/Adaptive_histogram_equalization) oraz do artykułu o metodzie CLAHE - Zuiderveld, Karel. “Contrast Limited Adaptive Histograph Equalization.” Graphic Gems IV. San Diego: Academic Press Professional, 1994. 474–485.

       Ponownie na podstawie rysunku proszę to przemyśleć.

6. "Kropka nad i" to oczywiście binaryzacja z wyznaczonym poprzez interpolację progiem - proszę dobrać rozmiar okna.
7. Na koniec proszę porównać na wspólnym rysunku poznane metody binaryzacji:
- Otsu,
- lokalna na podstawie średniej,
- lokalna Sauvoli,
- lokalna w oknach bez interpolacji,
- lokalna w oknach z interpolacją.

Proszę pod porównaniem, w osobnej sekcji *markdown*, krótko skomentować uzyskane wyniki.

```python
def local_binarization_with_interpolation(image_name, W):
    image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)
    height, width = image.shape

    # Tworzenie pustego obrazu docelowego dla wynikowej binaryzacji z interpolacją
    result_image = np.zeros((height, width), dtype=np.uint8)

    # Iteracja po obrazie i stosowanie binaryzacji lokalnej z interpolacją
    for i in range(0, height, W):
        for j in range(0, width, W):
            roi = image[i:i+W, j:j+W]  # Region zainteresowania (okno)
            
            # Binaryzacja adaptacyjna na regionie zainteresowania
            _, binary_region = cv2.threshold(roi, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
            
            # Interpolacja - przykładowo używając metody bilinearnej
            resized_binary_region = cv2.resize(binary_region, (W, W), interpolation=cv2.INTER_LINEAR)
            
            # Wstawienie przetworzonego regionu z powrotem do obrazu wynikowego
            result_image[i:i+W, j:j+W] = resized_binary_region

    return result_image

def local_binarization(image_name, W):
    image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)
    height, width = image.shape

    # Tworzenie pustego obrazu docelowego dla wynikowej binaryzacji lokalnej
    result_image = np.zeros((height, width), dtype=np.uint8)

    # Iteracja po obrazie i stosowanie binaryzacji lokalnej
    for i in range(0, height, W):
        for j in range(0, width, W):
            roi = image[i:i+W, j:j+W]  # Region zainteresowania (okno)
            
            # Binaryzacja adaptacyjna na regionie zainteresowania
            _, binary_region = cv2.threshold(roi, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
            
            # Wstawienie przetworzonego regionu z powrotem do obrazu wynikowego
            result_image[i:i+W, j:j+W] = binary_region

    return result_image


def show_results(image_name, W):
    image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)
    height, width = image.shape
    fig, axs = plt.subplots(2, 3, figsize=(15, 10))
    axs[0, 0].imshow(image, 'gray')
    axs[0, 0].set_title('Original')
    axs[0, 0].axis('off')
    axs[0, 1].imshow(cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1], 'gray')
    axs[0, 1].set_title('Otsu')
    axs[0, 1].axis('off')
    axs[0, 2].imshow(cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 2 * W + 1, 0), 'gray')
    axs[0, 2].set_title('Local mean')
    axs[0, 2].axis('off')
    axs[1, 0].imshow(cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 2 * W + 1, 0), 'gray')
    axs[1, 0].set_title('Local Sauvola')
    axs[1, 0].axis('off')
    axs[1, 1].imshow(local_binarization(image_name, W), 'gray')
    axs[1, 1].set_title('Binaryzacja lokalna bez interpolacji')
    axs[1, 1].axis('off')
    axs[1, 2].imshow(local_binarization_with_interpolation(image_name, W), 'gray')
    axs[1, 2].set_title('Interpolacja przy lokalnej binaryzacji')
    axs[1, 2].axis('off')
    plt.show()
    
show_results(fileName, 16)
```

## Wnioski
Metoda Otsu dobrze radzi sobie z globalną binaryzacją, lecz może być nieefektywna w przypadku obrazów z niskim kontrastem.

Binaryzacja lokalna na podstawie średniej działa dobrze wyróżniając lokalne detale, jednak może być wrażliwa na zmienne oświetlenie.

Binaryzacja lokalna Sauvoli radzi sobie z niwelowaniem problemu zmiennej jasności na obrazie.

Binaryzacja lokalna w oknach bez interpolacji traktuje każde okno niezależnie, co może prowadzić do kanciastych artefaktów w rezultacie.

Binaryzacja lokalna w oknach z interpolacją stara się wygładzić binaryzowane regiony, co może prowadzić do bardziej estetycznych wyników w porównaniu z wersją bez interpolacji. Jednakże, może również wprowadzić pewną utratę ostrości krawędzi.
